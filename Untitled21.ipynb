```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV3Large
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
import os
```

```python
import numpy as np
```

```python
import zipfile

z = zipfile.ZipFile("/content/drive/MyDrive/12 pfl/dataset.zip", "r")
z.extractall()
```

```python
path = "/content/dataset/train"
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
    )
train_generator = train_datagen.flow_from_directory(
    path,
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    subset="training"
)
validation_generator = train_datagen.flow_from_directory(
    path,
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    subset="validation"
)
model = MobileNetV3Large(weights="imagenet",input_shape=(224, 224, 3),include_top=False,classifier_activation="softmax")
x = model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation="relu")(x)
predict = Dense(5, activation="softmax")(x)
model = Model(inputs=model.input, outputs=predict)

model.compile(optimizer=Adam(learning_rate=0.0001), loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(validation_generator)

)
```

```python
test_dir = "/content/dataset/test"
```

```python
test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    labels=None,  # Нет меток для тестовых данных
    image_size=(224, 224),
    batch_size=32,
    shuffle=False
)
```

```python
predictions = model.predict(test_dataset, steps=len(test_generator))
danne = np.argmax(predictions, axis=1)
```
