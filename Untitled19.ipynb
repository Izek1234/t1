```python
# Функция для расчета процента точности
def calculate_accuracy(your_data, correct_data, tolerance=1e-6):
    # Количество правильных клиник
    correct_count = 0

    # Перебор всех клиник в вашем решении
    for i in range(len(your_data)):
        # Проверка, есть ли такая клиника в правильном решении
        for j in range(len(correct_data)):
            if np.allclose(your_data[i], correct_data[j], atol=tolerance):
                correct_count += 1
                break

    # Процент точности
    accuracy = (correct_count / len(your_data)) * 100
    return accuracy

# Пример использования
accuracy = calculate_accuracy(your_solution, correct_solution, tolerance=1e-6)
print(f"Процент точности: {accuracy:.2f}%")
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Загрузка данных
data = pd.read_csv('/content/data (1).csv')

# Параметры
alpha = 0.2
beta = 0.3
penalty_per_clinic = 0.122

# Функция для вычисления метрики T
def calculate_T(clinics, data, alpha, beta, penalty_per_clinic):
    distances_manhattan = cdist(data, clinics, metric='cityblock')
    distances_euclidean = cdist(data, clinics, metric='euclidean')
    min_distances = alpha * distances_manhattan + beta * distances_euclidean
    T = np.mean(np.min(min_distances, axis=1)) + len(clinics) * penalty_per_clinic
    return T

# Поиск оптимального количества клиник
best_T = float('inf')
best_clinics = None
best_k = 0

for k in range(1, 20):  # Попробуем от 1 до 20 клиник
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    clinics = kmeans.cluster_centers_
    T = calculate_T(clinics, data.values, alpha, beta, penalty_per_clinic)

    if T < best_T:
        best_T = T
        best_clinics = clinics
        best_k = k

# Сохранение результатов
np.savetxt('clinics1.txt', best_clinics, fmt='%.6f')

print(f"Оптимальное количество клиник: {best_k}")
print(f"Минимальное значение T: {best_T}")
```

```python
def calculate_score(T):
    # Константы
    T_max = 3.9892
    T_min = 3.9277

    # Вычисление score
    score = 25 + 50 * ((T_max - T) / (T_max - T_min))

    # Ограничение score в диапазоне [0, 100]
    score = max(0, min(100, score))

    return score

# Пример использования
T = 3.95  # Ваше значение метрики T
score = calculate_score(best_T)
print(f"Балл: {score:.2f}")
```
